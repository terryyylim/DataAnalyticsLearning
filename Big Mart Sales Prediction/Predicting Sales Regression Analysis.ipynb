## Problem Statement and Hypothesis Generation
The aim is to build a predictive model and find out the sales of each product at a particular store. Using this model, BigMart will try to understand the properties of products and stores which play a key role in increasing sales. From this, it is evident that this is a regression problem. We will be estimating the relationships between variables.

We will be testing based on the following hypothesis, for all variables.
H0: There is no significant prediction of sales figures by (variable).
H1: There is significant prediction of sales figures by (variable).

## Data Dictionary

| Outlet_Location_Type | The type of city in which the store is located |

## Data Exploration
Let's start off with understanding correlation between the variables in our dataset. Correlation is one of the most widely used statistical concepts, and it helps to describe the association between random variables.

<strong>1) Pearson Correlation Coefficient, ρ</strong><br>
This is the most widely used correlation coefficient. It measures the linear association between continuous variables.<br>

<li>The closer ρ is to 1, the more an increase in one variable associates with an increase in the other. On the other hand, the closer ρ is to -1, the increase in one variable would result in decrease in the other. Note that if X and Y are independent, then ρ is close to 0, but not vice versa!</li>
<li>The regression line through each scatter plot is optimal, as it minimizes the distance of all points to itself. As such, the slope of the line is mathematically equivalent to correlation between the 2 variables.</li>

#### Assumptions and Limitations
It is assumed that Pearson ρ assumes a linear relationship between the 2 variables. Pearson Correlation relies on means and standard deviations, meaning it is only defined for distributions where those statistics are finite, making the coefficient sensitive to outliers.<br> (Eg. ρ = -0.65 implies that (-0.652)*100 = 42% of variation in Y can be explained by X)

<strong>2) Spearman's Correlation Coefficient</strong>, returns a value between -1 and +1<br>
This is a special case of Pearson ρ. It assesses how well the relationship between 2 variables can be described using a monotonic function (whether linear or not). So, rather than comparing means and variances, it looks at the relative order of values for each variable.<br>

Spearman's Correlation Coefficient is appropriate for both continuous and discrete ordinal variables.

<li>Eg. A Spearman correlation of 1 results when the two variables being compared are monotonically related, even if their relationship is not linear. This means that all data-points with greater x-values than that of a given data-point will have greater y-values as well. In contrast, this does not give a perfect Pearson correlation.</li>

*Different formulation to be used depending if data has tied ranks or not

<strong>3) Kendall's Tau Coefficient, τ</strong>, returns a value between 0 and 1<br>
It does not take into account the difference between ranks, only directional agreement.<br>

Kendall’s Tau = (C – D / C + D),
where C is the number of concordant pairs and D is the number of discordant pairs.

*This is particularly useful for discrete data, where the relative position of data points is more important than difference between them

## Assumptions in Regression
Good results will not be delivered if the dataset does not fulfil the assumptions involved in regression analysis. Thus, we shall validate these assumptions.

<strong>1) Linear and Additive</strong><br>
There should be a linear and additive relationship between dependent variable and independent variable(s). A linear relationship suggests that a change in response Y due to one unit change in X is constant, regardless of the value of X. An additive relationship suggests that the effect of X on Y is independent of other variables.<br>

<strong>2) Autocorrelation</strong><br>
There should be no correlation between the residual (error) terms. Absence of this phenomenon is known as Autocorrelation.<br>

<strong>3) Multicollinearity</strong><br>
The independent variables should not be correlated. Absence of this phenomenon is known as multicollinearity. <br>

<strong>4) Heteroskedasticity</strong><br>
The error terms must have constant variance. This phenomenon is known as homoskedasticity. The presence of non-constant variance is referred to heteroskedasticity.<br>

```python
# independent variables
ind_train = final_df_train.drop('Item_Outlet_Sales', axis=1)

# Creating target values dataframe
df_target = df_train['Item_Outlet_Sales']

model = smf.OLS(df_target, ind_train)
model_fit = model.fit()
print(type(model_fit))
print(type(model_fit.get_influence()))

# fitted values (need a constant term for intercept)
model_fitted_y = model_fit.fittedvalues
#print(model_fitted_y)

# model residuals
model_residuals = model_fit.resid
#print(model_residuals)

# normalized residuals
model_norm_residuals = model_fit.get_influence().resid_studentized_internal

# absolute squared normalized residuals
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))

# absolute residuals
model_abs_resid = np.abs(model_residuals)

# leverage, from statsmodels internals
model_leverage = model_fit.get_influence().hat_matrix_diag

# cook's distance, from statsmodels internals
model_cooks = model_fit.get_influence().cooks_distance[0]

plot_model_1 = plt.figure(1)
plot_model_1.set_figheight(8)
plot_model_1.set_figwidth(10)

plot_model_1.axes[0] = sns.residplot(ind_train['Outlet_Size_cat'], df_target, lowess=True, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_model_1.axes[0].set_title('Residuals vs Fitted')
plot_model_1.axes[0].set_xlabel('Fitted values')
plot_model_1.axes[0].set_ylabel('Residuals')
plt.show()
```

## What is regression?
Regression is one way of modeling the strength and direction of the relationship between a dependent or output variable (usually represented by y) and one or more independent or input variables (usually represented by x). It differs from correlation analysis because it allows us to predict the outcome for new input or inputs we haven't seen before.

## What types of regression are there?
There are many different types of regression. The specific family of regressions we’ll be learning are called “generalized linear models”. The important thing for us to know is that with this family of models, we need to pick a specific type of regression we’re interested in. The type of regression will depend on what type of data we’re trying to predict.

<li> Linear: When we’re predicting a continuous value. (What temperature will it be today?) </li>
<li> Logistic: When we’re predicting which category our observation is in. (Is this is a cat or a dog?) What is regression? </li>
<li> Poisson: When we’re predicting a count value. (How many dogs will I see in the park?) </li>

### Practice Example: Recipe
For this practice, we will be attempting to predict whether or not a recipe is a dessert based on how many calories it has.

```python
# Required libraries
import numpy as np
import pandas as pd
from pandas.lib import Timestamp
from pandas import read_csv

# Plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns
from ggplot import *

# Creating recipes dataframe
df = pd.read_csv('epi_r.csv')
print(df)
```

## Cleaning of our dataset
Let us first do a scatterplot for 'dessert' and 'calories' attribute and try to find outliers. If there are outliers, we shall remove them.

```python
# Since 10000 calories seem way too high for a dessert, let's treat them as outliers
plt.scatter(df['calories'],df['dessert'])
plt.show()

# Let's see what items are listed to have more than 10000 calories
def find_outliers():
    length = len(df)
    for row in range(length):
        if (df.iloc[row]['calories'] > 10000):
            print(df.iloc[row]['title'])                          
find_outliers()

# Remove the rows with food items having more than 10000 calories
df = df.loc[df['calories'] < 10000]
```

Next, we want to remove any row of data with missing calorie value.

```python
cleaned_df = df.dropna(axis=0, subset=['calories'])
print(cleaned_df)
```

## Data Exploration
Now that we have cleaned our dataset, it's time to think what type of regression model we should be building for the task we want to accomplish.<br>
In this case, we are trying to predict whether a recipe is a dessert or not. <br>
As such, we are trying to predict a category for the recipe, i.e 'Is a dessert' and 'Is not a dessert'. <br><br>
Hence, we shall build a logistic regression model.

<strong>Linear Regression</strong>
<li>Family: Gaussian</li>
<li>Type of data: Continuous</li>

<strong>Logistic Regression</strong>
<li>Family: Binomial</li>
<li>Type of data: Categorical</li>

<strong>Poisson Regression</strong>
<li>Family: Poisson</li>
<li>Type of data: Count</li>

```python
# Plotting calories(independent variable) against categorical outcome(dependent variable)
# Method 1: Using ggplot
# ggplot(cleaned_df, aes(x='calories', y='dessert')) + geom_point() + scale_x_continuous(limits=(0,10000))

# Method 2: Using seaborn.regplot
sns.set(style="darkgrid")
print(type(cleaned_df['dessert'][0]))
print(type(cleaned_df['calories'][0]))
g = sns.regplot(x='calories',y='dessert', data=cleaned_df, logistic=True)
g.figure.set_size_inches(8, 8)
plt.show()
```

## Conclusion

That green line we've plotted is actually a regression model! Looking at the regression line, we can tell that as the number of calories in a recipe increases, it's more likely that that recipe is a dessert. (We can tell this because the line moves towards 1, which means "is a dessert" and away from 0, which means "isn't a dessert").

The grey shape behind it is the standard error. Looking at the error, we can tell that we're pretty sure about our prediction when there's fewer calories (towards the left side of the graph), but as we move to the right we get less certain and our error increases.
